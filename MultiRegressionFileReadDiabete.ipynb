```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
X=tf.placeholder(tf.float32,shape=[None ,8])
Y=tf.placeholder(tf.float32, shape=[None,1])
```

```python
xy= np.loadtxt('C:\교육\인공지능 강의_소프트웨어인재개발원\김성훈교수님\실습자료_이재오\diabetes.csv',delimiter=',', dtype=np.float32)
```

```python
x_data= xy[:,0:-1]
y_data= xy[:,[-1]]
```

```python
#두개의 traing 을 위한 변수 선언
W= tf.Variable(tf.random_normal([8,1]),name='weight')
b= tf.Variable(tf.random_normal([1]),name='bias')

hypothesis = tf.sigmoid(tf.matmul(X,W)+b)
```

```python
#cost/lost function
cost = -tf.reduce_mean(Y*tf.log(hypothesis) +(1-Y)*tf.log(1-hypothesis))
```

```python
#Minimize
Optimizer = tf.train.GradientDescentOptimizer(learning_rate =0.01)
train= Optimizer.minimize(cost)
```

```python
predict=tf.cast(hypothesis> 0.5,dtype=tf.float32)
accuracy =tf.reduce_mean(tf.cast(tf.equal(predict, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val,_= sess.run([cost,train],feed_dict={X:x_data,Y:y_data})
        if step %200 == 0:
            print(step,"COST:",cost_val)
    
    #Accuray Report
    h,c,a= sess.run([hypothesis,predict,accuracy],feed_dict={X:x_data,Y:y_data})
    print("\n Hypothesis:",h,"\nCorrect(Y):", c, "\nAccuracy:", a)

```

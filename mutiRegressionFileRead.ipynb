```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
X=tf.placeholder(tf.float32,shape=[None ,3])
Y=tf.placeholder(tf.float32, shape=[None,1])
```

```python
xy = np.loadtxt('C:\교육\인공지능 강의_소프트웨어인재개발원\김성훈교수님\실습자료_이재오\scoreData.csv', delimiter=',', dtype=np.float32)
print(xy)
```

```python
#두개의 traing 을 위한 변수 선언
W= tf.Variable(tf.random_normal([3,1]),name='weight')
b= tf.Variable(tf.random_normal([1]),name='bias')

hypothesis = tf.matmul(X,W)+b
```

```python
x_data=xy[:,0:-1]
y_data=xy[:,[-1]]
print(x_data)
print(y_data)
```

```python
#cost/lost function
cost = tf.reduce_mean(tf.square(hypothesis -Y))
```

```python
#미분을 이용하는 gradient decent를 이용하여 최소화
learning_rate =0.1
gradient = tf.reduce_mean((W*X -Y)*X) 
decent = W -learning_rate*gradient
update = W.assign(decent)
```

```python
#Minimize
Optimizer = tf.train.GradientDescentOptimizer(learning_rate =0.000001)
train= Optimizer.minimize(cost)
```

```python
sess =tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
W_val=[]
const_val=[]
```

```python
for step in range(2001):
    cost_val,hy_val,_= sess.run([cost,hypothesis,train],feed_dict={X:x_data,Y:y_data})
    if step %20 == 0:
        print(step,"COST:",cost_val,"prediction :", hy_val)
#plt.plot =(W_val,cost_val)
#plt.show()

```
